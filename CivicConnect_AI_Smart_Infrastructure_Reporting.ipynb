```python
import folium
from folium.plugins import LocateControl
import ipywidgets as widgets
from IPython.display import display, clear_output

# 1. Setup sample jurisdiction boundary boundaries (Town of Greenburgh focus)
def find_jurisdiction(lat, lon):
    # Greenburgh approximate bounding box for demo
    if 41.00 <= lat <= 41.08 and -73.88 <= lon <= -73.78:
        return "Town of Greenburgh DPW"
    elif 40.95 <= lat <= 41.15 and -73.95 <= lon [-73.70]:
        return "Westchester County Dept. of Public Works"
    else:
        return "New York State Dept. of Transportation (NYSDOT)"

# 2. Build Interactive UI Controls (Side Panel)
address_input = widgets.Text(
    value='Greenburgh, NY',
    placeholder='Enter an address or landmark',
    description='Location:',
    style={'description_width': 'initial'}
)

issue_type = widgets.Dropdown(
    options=['Pothole', 'Broken Sidewalk', 'Flooding', 'Missing Curb Ramp', 'Damaged Sign', 'Graffiti', 'Broken Streetlight'],
    value='Pothole',
    description='Issue Type:',
    style={'description_width': 'initial'}
)

description_input = widgets.Textarea(
    value='Deep pothole near the crosswalk causing traffic to slow down.',
    placeholder='Describe what you see...',
    description='Details:',
    style={'description_width': 'initial'}
)

submit_btn = widgets.Button(
    description='Submit Report',
    button_style='primary',
    icon='paper-plane'
)

output = widgets.Output()

# 3. Create Map centered on Greenburgh, NY
def generate_map(lat=41.033, lon=-73.824):
    m = folium.Map(location=[lat, lon], zoom_start=13, tiles="OpenStreetMap")

    # Add a marker at the chosen pinpoint
    folium.Marker(
        [lat, lon],
        popup="Reported Pinpoint",
        tooltip="Click to view pinpoint",
        icon=folium.Icon(color="red", icon="info-sign")
    ).add_to(m)

    return m

# 4. Handle Submission
def on_submit_clicked(b):
    with output:
        clear_output()
        # For demo: default coordinates for Greenburgh center
        lat, lon = 41.033, -73.824
        agency = find_jurisdiction(lat, lon)

        print("✅ REPORT SUBMITTED SUCCESSFULLY!")
        print("-----------------------------------")
        print(f"📍 Location: {address_input.value} ({lat}, {lon})")
        print(f"⚠️ Category: {issue_type.value}")
        print(f"📝 Description: {description_input.value}")
        print(f"🏛️ CivicConnect AI Routing: Assigned to {agency}")

submit_btn.on_click(on_submit_clicked)

# 5. Display Form & Map Side-by-Side
form_box = widgets.VBox([
    widgets.HTML("<h3>📍 CivicConnect AI - Report an Issue</h3>"),
    address_input,
    issue_type,
    description_input,
    submit_btn,
    output
])

display(form_box)
display(generate_map())
```
